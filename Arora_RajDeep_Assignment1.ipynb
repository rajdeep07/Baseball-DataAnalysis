### Q13 : Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number)  

```python
#grouping data by yearID
groupedByYear=dat.groupby(['yearID','nameFirst','nameLast','playerID'],as_index=False).sum()

# getting list of years
seasons=groupedByYear['yearID'].unique()

playername= []
for year in seasons:
    maxhits=groupedByYear[groupedByYear['yearID']==year]['H'].max()
    playername.append(groupedByYear[(groupedByYear['yearID']==year) & (groupedByYear['H']==maxhits)]['playerID'].get_values())

#flattening list of players    
import itertools
flattened_list  = list(itertools.chain(*playername))

#counting 
from collections import Counter
sortedList=Counter(flattened_list).most_common()

hitsLeader=pd.DataFrame(sortedList,columns=['playerID','frequency'])
hitsLeader.head(3)
```

### Q24 : Who is the youngest player to hit 50 or more HRs in a single season? (Dataframe)

### Q25 : Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe)

```python
dat['PlayingAge'] = dat['yearID'] - dat['birthYear']
dat.dropna()
dat1 = dat.groupby(["playerID","PlayingAge"]).sum()['HR'].groupby(["playerID"]).cumsum().reset_index()
dat1[dat1['HR']>=300].sort_values(['PlayingAge'],ascending = [1]).head()
```
